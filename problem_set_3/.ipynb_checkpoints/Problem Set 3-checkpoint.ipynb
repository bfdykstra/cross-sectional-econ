```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from __future__ import division
import seaborn as sns

%matplotlib inline
import matplotlib.pylab as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 18.5, 10.5
```

```python
def print_resids(preds, resids):
    ax = sns.regplot(preds, resids);
    ax.set(xlabel = 'Predicted values', ylabel = 'errors', title = 'Predicted values vs. Errors')
    plt.show();
```

```python
fertility = pd.read_stata("http://rlhick.people.wm.edu/econ407/data/fertility.dta")
```

Theres a huge problem with missing data in this dataset, roughly a quarter of agefbrth data a missing from the total amount of observations. I get the data now that has only not null values for age of first birth, education, tv, usemeth and urban.

```python
no_null = fertility[(fertility['agefbrth'].notnull()) & (fertility['educ'].notnull()) & 
                    (fertility['yearborn'].notnull()) & (fertility['electric'].notnull()) 
                    & (fertility['ceb'].notnull())] #gets all columns that aren't null

print "lost {} samples of data".format(fertility.shape[0] - no_null.shape[0])

ind_vars = ['educ', 'yearborn', 'electric', 'ceb']
dep_var = 'agefbrth'
x = no_null[ind_vars] 
y = no_null[dep_var]
```

```python
x_const = sm.add_constant(x)

first_model_results = sm.OLS(y, x_const).fit()

#results = first_model.fit()

print first_model_results.summary()
```

```python
print_resids(first_model_results.predict(x_const), first_model_results.resid)
```

This is some of the worst error structure I have seen
